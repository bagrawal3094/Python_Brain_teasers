```python
# 1) reverse of a string and check if it's a palindrome
# 2) Find out all the Palindromes in a string
# 3) Find out the min # of letters to add in a string to make it a Palindrome

import collections as cs

# 1

def checkPalindrome(s):
    
    temp = ''
    
    for i in range(len(s)):
        
        temp = temp + (s[len(s)-i-1])
        
    return s== temp

checkPalindrome('bharat')

# 2) find out all Palindrome in a string

def all_palindrome(check_s):
    
    result = []
    for i in range(len(check_s)):

        first = check_s[i]
        
        for j in range(i+1, len(check_s)):

            temp_s = check_s[i:j]

            if checkPalindrome(temp_s) and len(temp_s)>1:
                result.append(temp_s)

    return result

check_s = 'abccbabcddcbsdfasdsgdlkad'

all_palindrome(check_s)

# 3) min # of  letters need to add to make a string palindrome
def min_pal_s(pal_s):
    
    cnt = cs.Counter()
    for each in pal_s:

        cnt[each] +=1
    
    pal_dict = dict(cnt)

    min_count = 0
    for key, val in pal_dict.items():
        
        
        if val % 2 == 1:
            min_count += 1
    
    return min_count -1
    
pal_s = 'bharat'

min_pal_s(pal_s)
```
